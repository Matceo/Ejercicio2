```python
#Ejercicio1

#Caso 1: t<0.

import sympy as sym

class rect(sym.Function):
    @classmethod
    def eval(cls, arg):
        return sym.Heaviside(arg + sym.S.Half) - sym.Heaviside(arg - sym.S.Half)

sym.init_printing()
t, tau = sym.symbols('t tau', real=True)
x = rect(t - 1/2)
h = sym.exp(-t) * sym.Heaviside(t)
h1 = h.subs(t, tau)
h2 = h1.subs(tau, -tau)
h3 = h2.subs(tau, tau-t)
x1 = x.subs(t, tau)
y=sym.integrate(x1*h3,(tau,-100,0))
sym.plot(y,(t,-15,15),ylim=[0,1])
```

```python
#Caso 2: 0<=t<1.

import sympy as sym
y=sym.integrate(x1*h3,(tau,0,1))
sym.plot(y,(t,-15,15),ylim=[0,1])
```

```python
#Caso 3: t>1.

import sympy as sym
y=sym.integrate(x1*h3,(tau,1,100))
sym.plot(y,(t,-15,15),ylim=[0,1])
```
